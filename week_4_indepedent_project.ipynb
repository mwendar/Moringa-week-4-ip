What postal code is the most popular for picking up Blue cars? Does the most popular station belong to that postal code?
Overall?

```python
#Create a list that will store a postal code after the loop if it has a larger Bluecar counter than the previous one
popularPostalCode = []
#Variable to store the highest value of the Bluecar counter sum per Postal Code
popularSum = 0
#Get a list of the unique postal codes and loop through each unique postal code

for pc in df['Postal code'].unique():
  #Display each postal code against all its Blue counter values
  print(df.loc[df['Postal code']==pc, ['Postal code', 'Bluecar counter']])
  #Store the above dataframe in a variable df2
  df2 = df.loc[df['Postal code']==pc, ['Postal code', 'Bluecar counter']]
  #Count the sum of each unique postal codes' Bluecar counter and print
  print(df2['Bluecar counter'].sum())
  currentSum = df2['Bluecar counter'].sum()#Store the sum in a variable df2
  #With an if statement, check for the highest value of the sum and store its postal code and sum
  
  if currentSum > popularSum:
    popularSum = currentSum
    #Add the larger value at the end of the list
    popularPostalCode.append(pc)

#Print the last postal code from the list and the highest sum stored
print("The most popular code is " + str(popularPostalCode[-1]) + " with a frequency of " + str(popularSum))
```
